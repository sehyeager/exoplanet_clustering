```python
#this drops 10 columns from our dataframe.  Not a terrible loss!  
#From here we can imput our data better.
df = df.dropna(axis = 1, thresh = 2176)
```

```python
df.isna().sum()
```

```python
#testing a thorough null filling method:  fill with previous values up to a point, 
#and then fill with the average
#if I have to, I'll return to this to clean it up, change parameters, etc
df = df.fillna(method = 'bfill', limit =25)
df = df.fillna(value = df.median())
```

```python
df.isna().sum()
```

```python
df.to_csv('./backfill_mean_data.csv', index = False)
```
